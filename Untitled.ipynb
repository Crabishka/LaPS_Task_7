```python
import matplotlib.pyplot as plt 
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
```

# 

```python
u,v = np.mgrid[0:math.pi*2:200j, -1:1:200j]
```

```python
r = 1 + (v/2) * np.cos(u/2)
x = r * np.cos(u)
y = r * np.sin(u)
z = (v/2) * np.sin(u/2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, cmap='inferno')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
```
